```python
# Load yfinance table from MySQL
df_yfinance = pd.read_sql('''select * from yfinance where date between "2020-04-14" and "2020-05-15"
                          ORDER BY date asc''', con=engine)
df_yfinance.rename(columns = {'Date':'date'}, inplace=True)
df_yfinance.rename(columns = {'percentage_change': 'percent_change'}, inplace = True)
df_yfinance['date'] = df_yfinance['date'].apply(pd.to_datetime)
df_yfinance.set_index('date', inplace = True)
df_yfinance
```

```python
# Load news table from MySQL
df_news = pd.read_sql('select * from news where date between "2020-04-14" and "2020-05-15" ', con=engine)
df_news['date'] = df_news['date'].apply(pd.to_datetime)
df_news.set_index('date', inplace = True)
df_news
```

```python
# Load sentiment table from MySQL
df_sentiment = pd.read_sql('select * from sentiment where date between "2020-04-14" and "2020-05-15" ', con=engine)
df_sentiment['date'] = df_sentiment['date'].apply(pd.to_datetime)
df_sentiment.rename(columns = {'daily_score':'daily_sentiment_score'}, inplace=True)
df_sentiment.set_index('date', inplace = True)
df_sentiment
```

```python
# Load twitter table from MySQL
df_twitter = pd.read_sql('select * from twitter where date between "2020-04-14" and "2020-05-15" ', con=engine)
df_twitter['date'] = df_twitter['date'].apply(pd.to_datetime)
df_twitter.set_index('date', inplace = True)
df_twitter
```

```python
# Load headline table from MySQL
pd.set_option('display.max_colwidth', -1)
df_headlines = pd.read_sql('select * from headlines', con=engine)
df_headlines.set_index('date', inplace = True)
df_headlines
```

```python
# Merge dfs
df_merged = df_yfinance.merge(df_news, left_on='date', right_on='date')
df_merged = df_merged.merge(df_sentiment, left_on='date', right_on='date')
df_merged = df_merged.merge(df_twitter, left_on='date', right_on='date')
#df_merged = df_merged.rename({'percentage_change': 'percent_change'}, axis=1)
df_merged = df_merged.rename({'count': 'tweet_count'}, axis=1)
df_merged
```

### Regression

```python
# Keep only relevant colunns
df_merged = df_merged.filter(['percent_change','total_articles','daily_sentiment_score', 'tweet_count'], axis=1)
df_merged
```

```python
# Compute correlation matrix
corr_matrix = df_merged.corr()
corr_matrix
```

```python
#pip install statsmodels
import statsmodels.formula.api as sm
```

```python
result = sm.ols(formula="percent_change ~ total_articles + daily_sentiment_score + tweet_count", data=df_merged).fit()
```

```python
print(result.summary())
```

```python
print(result.params)
```

### Regression using abs value

```python
# Use abs value of percent_change and daily_sentiment_score
df_merged2 = df_merged.copy()
df_merged2['percent_change'] = abs(df_merged2['percent_change'])
df_merged2['daily_sentiment_score'] = abs(df_merged2['daily_sentiment_score'])
df_merged2
```

```python
# Compute correlation matrix
corr_matrix2 = df_merged2.corr()
corr_matrix2
```

```python
from bokeh.io import output_notebook, show
from bokeh.models import HoverTool, DataRange1d
from bokeh.plotting import figure
from datetime import datetime
from bokeh.models.formatters import DatetimeTickFormatter
from bokeh.layouts import row

from bokeh.layouts import column, grid 

p3 = figure(x_axis_type='datetime', plot_width=500, plot_height=500)
p3b = figure(x_axis_type='datetime', plot_width=500, plot_height=500)

source = ColumnDataSource(data=dict(
        date=df_yfin['date'],
        #dollar_change=df_yfin['dollar_change'],
        dollar_change=df_yfin['Difference'],
        percent_change=df_yfin['percent_change'],

)
)

p3.line(x='date', y='dollar_change', source=source, color="navy")
p3.cross(x='date', y='dollar_change', fill_color='white', size=10, source=source)
p3b.line(x='date', y='percent_change', source=source, color="green")
p3b.cross(x='date', y='percent_change', fill_color='white', size=10, source=source)


hover = HoverTool(
tooltips=[
    ('Date', '@date{%m-%d-%Y}'),
    ('Dollar Change', '@dollar_change{%0.000000f}'),
      ],
    formatters={
        'date':'datetime',
        'dollar_change' : 'printf',
},
)
hover_b = HoverTool(
tooltips=[
    ('Date', '@date{%m-%d-%Y}'),
    ('Percent Change', '@percent_change{%0.00f}'),
      ],
    formatters={
        'date':'datetime',
#         'percent_change' : 'printf',
},
)
p3.add_tools(hover)
p3b.add_tools(hover_b)

p3.xaxis.formatter=DatetimeTickFormatter(
minutes=['%M'],
hours=['%H:%M'],
days=['%m/%d/%Y'],
months=["m/%Y"],
years=['%Y'],
)

p3b.xaxis.formatter=DatetimeTickFormatter(
minutes=['%M'],
hours=['%H:%M'],
days=['%m/%d/%Y'],
months=["m/%Y"],
years=['%Y'],
)

p3.title.text = 'Day to Day Stock Price Change'
p3.xaxis.axis_label = 'Date'
p3.yaxis.axis_label = '$ amount change'

p3b.title.text = 'Day to Day Stock Price % Change'
p3b.xaxis.axis_label = 'Date'
p3b.yaxis.axis_label = '% change'

layout = row(p3, p3b)
output_file("price_change.html")
output_notebook()


show(layout)


```
